```python
# Naive Bayes Done
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

#importing the dataset
dataset = pd.read_csv('Book2.csv')
print(dataset)

#result of independent variable, these are: age  income student credit_rating
X = dataset.iloc[:,0:4].values
print("\n Result of independent variable age, income, student, credit_rating = X:\n",X,"\n")

#result of dependent variable, i:e buys_computer
y = dataset.iloc[:, 4].values
print("\n buys_computer = y = ",y,"\n")

from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:,0]= labelencoder_X.fit_transform(X[:,0])
print("\n",X,"\n")
X[:,1]= labelencoder_X.fit_transform(X[:,1])
print("\n",X,"\n")
X[:,2]= labelencoder_X.fit_transform(X[:,2])
print("\n",X,"\n")
X[:,3]= labelencoder_X.fit_transform(X[:,3])
print("\n",X,"\n")

print ("Shape of the data:",dataset.shape,"\n")
# (14, 5)

X = np.array(X)
y = np.array(y)

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 10)

model = GaussianNB()
model.fit(X_train,y_train)
predicted1 = model.predict(X_test)
print("Prediction Result: ",predicted1)

print("Accuracy: ",accuracy_score(y_test, predicted1, normalize = True)) #accuracy_score(train output, predict output)

predicted2= model.predict([[2,2,1,1]])
print(predicted2)
```

```python
# Support Vector Machines Done
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score

#importing the dataset
dataset = pd.read_csv('Book2.csv')
print(dataset)

#result of independent variable, these are: age  income student credit_rating
X = dataset.iloc[:,0:4].values
print("\n Result of independent variable age, income, student, credit_rating = X:\n",X,"\n")

#result of dependent variable, i:e buys_computer
y = dataset.iloc[:, 4].values
print("\n buys_computer = y = ",y,"\n")

from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:,0]= labelencoder_X.fit_transform(X[:,0])
print("\n",X,"\n")
X[:,1]= labelencoder_X.fit_transform(X[:,1])
print("\n",X,"\n")
X[:,2]= labelencoder_X.fit_transform(X[:,2])
print("\n",X,"\n")
X[:,3]= labelencoder_X.fit_transform(X[:,3])
print("\n",X,"\n")

print ("Shape of the data:",dataset.shape,"\n")
# (14, 5)

X = np.array(X)
y = np.array(y)

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 10)

model = svm.SVC()
model.fit(X_train,y_train)
predicted1 = model.predict(X_test)
print("SVM Result:\n",predicted1)

print("Accuracy: ",accuracy_score(y_test, predicted1, normalize = True)) #accuracy_score(train output, predict output)

predicted2= model.predict([[2,2,1,1]])
print(predicted2)
```
